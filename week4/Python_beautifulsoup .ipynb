```python
t = map(lambda s: s.strip(), fin_names)
telimler = list(t)
print(telimler)
```

```python
url = 'http://www.qss.az/trainings'
response = requests.get(url)
trainings_soup =  BeautifulSoup(response.text, 'lxml')
#print(trainings_soup.prettify())
trainings_soup
```

```python
all_names,all_links, all_hours, all_lang = [],[], [], []

for product in trainings_soup.find_all('div',attrs={"news-title"}):
    print(product.find('a').get('href'))
    if(product.find('a').get('href')=='http://dsa.az'):
        continue
    else:
        link = 'http://www.qss.az' + product.find('a').get('href')
        names = product.find('a').text
        r2 = requests.get(link)
        c = r2.content
        soup2=BeautifulSoup(c,"html.parser")
        hours = soup2.find('h4', attrs={"single-hour"}).text
        language = soup2.find('h4', {'class': 'single-language'}).text
        #creation = product.find('div', {'class': 'products-created'}).text

    all_links.append(link)
    all_hours.append(hours)
    all_lang.append(language)
    all_names.append(names)


data_list = list(zip(all_names,all_links, all_hours, all_lang))
column_names = ['Names','Links', 'Houur','Language']
qss = pd.DataFrame(data_list,columns=column_names)
```

```python
print(qss)
```

```python
qss['Houur'] = qss['Houur'].str[8:]
```

```python
qss['Language'] = qss['Language'].str[5:]
```
